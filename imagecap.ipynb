```python
import pygame
import pygame.camera
import time
from PIL import Image



pygame.camera.init()
msg = pygame.camera.list_cameras() #Camera detected or not

print(msg)

cam = pygame.camera.Camera(msg[0],(640,480))
```

```python
cam.start()

time.sleep(0.1)

img = cam.get_image()
pygame.image.save(img,"filename.jpg")
#Load the image
img_open = Image.open('filename.jpg')
img_open
```
